```python
import pandas as pd
```

1. Show first five rows:

```python
df = pd.read_csv("week2-ds.tsv", sep = '\t')
df.head()
```

2. Number of Rows

```python
df.count()

length = len(df)
print(length)
```

3. Number of Distinct Rows

```python
unique_rows = len(df.drop_duplicates())
print(unique_rows)
```

4. Number of Distinct 'Hashed Device ID'

```python
unique_hash = len(df['Hashed Device ID'].drop_duplicates())
print(unique_hash)
```

5. Number of Distinct Columns

```python
unique_id = len(df['Polygon ID'].drop_duplicates())
print(f'Distinct Polygon IDs: {unique_id}')

unique_source = len(df['Data Source'].drop_duplicates())
print(f'Distinct Data Sources: {unique_source}')

unique_date = len(df['Date'].drop_duplicates())
print(f'Distinct Dates: {unique_date}')
```

6a. Date , Number of distinct Hashed Device IDs

```python
df.groupby('Date')['Hashed Device ID'].nunique().reset_index(name='Hashed Device ID')
```

6b. Polygon ID, Number of Distinct Hashed Device IDs

```python
df.groupby('Polygon ID')['Hashed Device ID'].nunique().reset_index(name='Hashed Device ID')
```

6c. Polygon ID, Date, Number of Distinct Hashed Device IDs


```python
df.groupby(['Polygon ID', 'Date'])['Hashed Device ID'].nunique().reset_index(name='Hashed Device ID')
```

```python
filter = source_x.groupby('Polygon ID')['Hashed Device ID'].nunique().reset_index(name='Hashed Device ID')
filter['Data Source'] = 'X'

filter
```

```python
filter = source_x.groupby(['Polygon ID', 'Date'])['Hashed Device ID'].nunique().reset_index(name='Hashed Device ID')
filter['Data Source'] = 'X'

filter
```
